```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import copy
```

```python
df_x_train = pd.read_excel('PS1_data.xlsx', sheet_name='X_train')
df_y_train = pd.read_excel('PS1_data.xlsx', sheet_name='Y_train')
```

```python
x_train = df_x_train.values.flatten()
y_train = df_y_train.values.flatten()
print("Data has been loaded from Excel successfully.")
```

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    total_cost = np.sum(((w * x + b) - y) ** 2) / (2 * m)
    return total_cost
```

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]
    error = (w * x + b) - y
    dj_dw = np.sum(error * x) / m
    dj_db = np.sum(error) / m
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters):
    w = copy.deepcopy(w_in)
    b = b_in
    J_history = []

    for i in range(num_iters):
        dj_dw, dj_db = gradient_function(x, y, w, b)
        w -= alpha * dj_dw
        b -= alpha * dj_db
        cost = cost_function(x, y, w, b)
        J_history.append(cost)
        if i % (num_iters // 10) == 0:
            print(f"Iteration {i}: Cost {cost:.2f}")

    return w, b, J_history
```

```python
initial_w = 0.
initial_b = 0.
iterations = 1500
alpha = 0.01
w, b, _ = gradient_descent(x_train, y_train, initial_w, initial_b, compute_cost, compute_gradient, alpha, iterations)
print("w, b found by gradient descent:", w, b)
```

```python
plt.scatter(x_train, y_train, marker='x', c='r', label='Training data')
plt.plot(x_train, w * x_train + b, label='Linear regression')
plt.title("Profits vs. Population per city")
plt.ylabel('Profit in $10,000')
plt.xlabel('Population of City in 10,000s')
plt.legend()
plt.show()
```

```python
predict1 = w * 3.5 + b
predict2 = w * 7.0 + b
print(f'For population = 35,000, we predict a profit of ${predict1 * 10000:.2f}')
print(f'For population = 70,000, we predict a profit of ${predict2 * 10000:.2f}')
```
